```python
X = df ['x'].values
Y = df ['y'].values
```

```python
plt.scatter (X,Y)
plt.show()
```

```python
#Importing Science Python Library
from scipy import stats
```

```python
slope, intercept, r, p, std_err = stats.linregress (X,Y)
```

```python
print(slope, intercept, r, p, std_err)
```

```python
plt.scatter (X,Y)
y = slope * X + intercept
plt.plot (X,y)
plt.show()
```
